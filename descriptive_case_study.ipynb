### Matplotlib and seaborn is a library used for visualization i.e. charts

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.set(font_scale=1.5)
```

```python
# Read in TN middle school dataset from GitHub
df = pd.read_csv('https://raw.githubusercontent.com/LearnDataSci/article-resources/master/Essential%20Statistics/middle_tn_schools.csv')
df.head(2)
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
sr_rl = df[['reduced_lunch', 'school_rating']]
```

```python
sr_rl.head(5)
```

```python
sr_rl_cat = sr_rl.groupby(['school_rating'])
```

```python
sr_rl_cat.describe()
```

```python
sr_rl.corr()
```

```python
fig, ax = plt.subplots(figsize=(14,7))
ax.set_ylabel('school_rating')

# boxplot with only these two variables
sr_rl.boxplot(by='school_rating', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

```python
plt.figure(figsize=(14,8)) # set the size of the graph
_ = sns.regplot(data=sr_rl, x='reduced_lunch', y='school_rating')
```

## Create tabular correlation matrix

```python
corr = df.corr()
_, ax = plt.subplots(figsize=(13,10)) 

# graph correlation matrix
_ = sns.heatmap(corr, ax=ax,
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values,
                cmap='coolwarm')
```
